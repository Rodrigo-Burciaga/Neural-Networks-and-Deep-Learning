```
import numpy as np
from math import pow
```

```
w = np.random.rand(1)
b = np.random.rand(1)

x = 5
y = 0
eta = .1
epochs = 100


def sigmoid(z):
  return 1.0/(1.0+np.exp(-z))

def sigmoid_prime(z):
    """"Derivative of the sigmoid function."""
    return sigmoid(z) * (1 - sigmoid(z))

def cost(a,y):
  return pow(y-a,2)

print(sigmoid(5))
  

for i in range(epochs):
  z = np.dot(w,x) + b
  a = sigmoid(z)
  delta = a - y * sigmoid_prime(z)
  b = b - eta * delta
  w = w - eta * (delta * x)
  costo = cost(a,y)
  print('costo ', costo)
  print(a, w, b)

 
```
